```python
import sqlite3 as sql

database ='./Database.db' # Creating a swl llite database connection 
connection_obj=sql.connect(database)

# convert the dataframe to sql db 
sales.to_sql('sales',connection_obj,if_exists='replace',index=False)

# SQL Sample code 
sql_code =  """ 
select 
    distinct city          
from sales 
order by city 
"""
# Cursor Object 
cursor_obj=connection_obj.cursor()
cursor_obj.execute(sql_code) # executes the code 
results = cursor_obj.fetchall()

df=pd.DataFrame(results,columns=['city'])
df
```

# Answer Here

```python
# Question 1 : List all unique ship modes 

sql_code =  """ 
SELECT
    DISTINCT [Ship Mode]  as ship_mode       
FROM sales 
"""
# Cursor Object 
cursor_obj=connection_obj.cursor()
cursor_obj.execute(sql_code) # executes the code 
results = cursor_obj.fetchall()

df=pd.DataFrame(results,columns=['Ship Modes']) 
df
```

```python
# Question 2 : List all unique ship modes 

sql_code =  """ 
SELECT
    DISTINCT [Ship Mode]  as ship_mode       
FROM sales 
"""
# Cursor Object 
cursor_obj=connection_obj.cursor()
cursor_obj.execute(sql_code) # executes the code 
results = cursor_obj.fetchall()

df=pd.DataFrame(results,columns=['Ship Modes']) 
df
```

```python

# Question 3 : Count the total number of orders.

sql_code =  """ 
SELECT
    COUNT (DISTINCT [Order ID])  AS no_of_orders     
FROM sales 
"""
# Cursor Object 
cursor_obj=connection_obj.cursor()
cursor_obj.execute(sql_code) # executes the code 
results = cursor_obj.fetchall()

df=pd.DataFrame(results,columns=['no_of_orders']) 
df
```

```python

# Question 4 : Find the earliest order date.

sql_code =  """ 
SELECT
    MAX([Order Date])  AS no_of_orders     
FROM sales 
"""
# Cursor Object 
cursor_obj=connection_obj.cursor()
cursor_obj.execute(sql_code) # executes the code 
results = cursor_obj.fetchall()

df=pd.DataFrame(results,columns=['Latest_order_date']) 
df
```

```python
# Question 5 : Retrieve the total sales amount.

sql_code =  """ 
SELECT
    round(SUM(sales)/1000000,2)  AS total_sales_amount_in_millions     
FROM sales 
"""
# Cursor Object 
cursor_obj=connection_obj.cursor()
cursor_obj.execute(sql_code) # executes the code 
results = cursor_obj.fetchall()

df=pd.DataFrame(results,columns=['Total_Sales_Millions']) 
df
```
